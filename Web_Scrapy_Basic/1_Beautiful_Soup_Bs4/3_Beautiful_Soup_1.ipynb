```python
import json
from datetime import date
import urllib
from bs4 import BeautifulSoup
import time

today = str(date.today())



#web_page = input('Enter - ') #we going to use https://www.bet365.com/#/AC/B13/C20720596/D8/E95739685/F8/
web_page = 'https://www.bet365.com/#/AC/B13/C20720604/D7/E83/F4/'

#print(soup.prettify())

thisdict = {}
for x in range(100):
    #print(x)
    web_code = urllib.request.urlopen(web_page)
    soup = BeautifulSoup(web_code,"html.parser")
    check = 'soup' + str(x)
    thisdict[check]=soup.decode('utf-8') #without decode I was having problems with " Is not JSON serializable"
    time.sleep(1)

file = 'bet365_' + today + '.json'  #instead File_2020-11-18 , we want File_2020_11_18
File_out=file.replace('-','_')

with open(File_out, 'w') as fp:
    json.dump(thisdict, fp)
```
